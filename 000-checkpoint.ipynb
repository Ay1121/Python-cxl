```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from pathlib import Path
import os

# ======================  ======================
# 
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False

# 1. 
desktop_path = Path(os.path.expanduser("~")) / "Desktop"
main_output_dir = desktop_path / ""
figures_dir = main_output_dir / "figures"
os.makedirs(figures_dir, exist_ok=True)
print(f"{main_output_dir}")

# 2. raw
raw_data_dir = Path(r"C:\Users\ay\Desktop\raw")
main_data = pd.read_csv(raw_data_dir / "main_data_advanced.csv")
od_matrix = pd.read_csv(raw_data_dir / "od_matrix.csv")

# 2023
data_2023 = main_data[main_data[""] == 2023].copy()
years = sorted(main_data[""].unique())


# ======================  ======================
def partial_correlation(x, y, covariate):
    """"""
    # 
    slope_x, intercept_x, _, _, _ = stats.linregress(covariate, x)
    res_x = x - (slope_x * covariate + intercept_x)  # 
    
    slope_y, intercept_y, _, _, _ = stats.linregress(covariate, y)
    res_y = y - (slope_y * covariate + intercept_y)  # 
    
    # Pearson
    return stats.pearsonr(res_x, res_y)


# ====================== 5.1  ======================
print("=== 5.1  ===")

# ---------- 5.1.1  ----------
print("...")
# GB
yearly_transfer = main_data.groupby("")["_TB"].sum() * 1024
# 
yearly_trade = main_data.groupby("")["_"].sum()

# 
plt.figure(figsize=(12, 6))
plt.plot(yearly_transfer.index, yearly_transfer.values, 'o-', color='darkblue', linewidth=2, markersize=8, label='')
# 3
top_cities_2023 = data_2023.nlargest(3, "_TB")[""].tolist()
for city in top_cities_2023:
    city_data = main_data[main_data[""] == city].set_index("")["_TB"] * 1024
    plt.plot(city_data.index, city_data.values, 'o-', linewidth=1.5, markersize=6, label=city)
plt.title('2019-2023', fontsize=15)
plt.xlabel('', fontsize=12)
plt.ylabel('GB', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()

# 
plt.figure(figsize=(12, 6))
plt.plot(yearly_trade.index, yearly_trade.values, 'o-', color='darkred', linewidth=2, markersize=8, label='')
top_trade_cities_2023 = data_2023.nlargest(3, "_")[""].tolist()
for city in top_trade_cities_2023:
    city_data = main_data[main_data[""] == city].set_index("")["_"]
    plt.plot(city_data.index, city_data.values, 'o-', linewidth=1.5, markersize=6, label=city)
plt.title('2019-2023', fontsize=15)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ---------- 5.1.2  ----------
print("...")
# 
city_coords = {
    "": (113.27, 23.13), "": (114.07, 22.62), "": (113.56, 22.27),
    "": (113.12, 23.02), "": (114.41, 23.09), "": (113.75, 22.94),
    "": (113.38, 22.52), "": (113.09, 22.59), "": (112.47, 23.05),
    "": (114.17, 22.28), "": (113.54, 22.19)
}
# 
cities = data_2023[""].tolist()
x = [city_coords[city][0] for city in cities]
y = [city_coords[city][1] for city in cities]
transfer_values = data_2023["_TB"].values
sizes = transfer_values / transfer_values.max() * 300  # 

# 
plt.figure(figsize=(12, 8))
sc = plt.scatter(x, y, s=sizes, alpha=0.7, c=transfer_values, cmap='viridis')
for i, city in enumerate(cities):
    plt.text(x[i], y[i], city, fontsize=9)
plt.colorbar(sc, label='TB')
plt.title('2023', fontsize=15)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ---------- 5.1.3  ----------
print("...")
# 
def gini_coefficient(x):
    x = np.sort(x)
    n = len(x)
    return (2 * np.sum(np.arange(1, n+1) * x) / np.sum(x) - (n + 1)) / n
# 
def theil_index(x):
    x = np.array(x)[np.array(x) > 0]
    n = len(x)
    mean_x = np.mean(x)
    return np.sum(x / mean_x * np.log(x / mean_x)) / n

# 
gini_values = []
theil_values = []
for year in years:
    year_data = main_data[main_data[""] == year]["_TB"].values
    gini_values.append(gini_coefficient(year_data))
    theil_values.append(theil_index(year_data))

# 
fig, ax1 = plt.subplots(figsize=(12, 6))
color1 = 'tab:blue'
ax1.set_xlabel('')
ax1.set_ylabel('', color=color1)
ax1.plot(years, gini_values, 'o-', color=color1, label='')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('', color=color2)
ax2.plot(years, theil_values, 's-', color=color2, label='')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('2019-2023', fontsize=15)
fig.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ====================== 5.2  ======================
print("\n=== 5.2  ===")

# ---------- 5.2.1  ----------
print("...")
# 
key_indicators = [
    '_TB', '_', '', '5G_',
    'GDP_', 'GDP_', '_', '_',
    '', ''
]
corr_data = data_2023[key_indicators].copy()
corr_matrix = corr_data.corr()

# 
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f", cmap='coolwarm',
            xticklabels=key_indicators, yticklabels=key_indicators,
            vmin=-1, vmax=1, center=0, square=True, linewidths=.5)
plt.title('2023', fontsize=15)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ---------- 5.2.2  ----------
print("...")
# 
x = data_2023['_TB']
y = data_2023['']
covariate = data_2023['GDP_']

# 
partial_corr, p_value = partial_correlation(x, y, covariate)

# 
plt.figure(figsize=(10, 6))
# 
slope_x, intercept_x = stats.linregress(covariate, x)[:2]
res_flow = x - (slope_x * covariate + intercept_x)

slope_y, intercept_y = stats.linregress(covariate, y)[:2]
res_innov = y - (slope_y * covariate + intercept_y)

plt.scatter(res_flow, res_innov, alpha=0.7)
plt.plot(np.unique(res_flow), np.poly1d(np.polyfit(res_flow, res_innov, 1))(np.unique(res_flow)), 'r--')
plt.title(f'GDP (r={partial_corr:.2f}, p={p_value:.3f})', fontsize=14)
plt.xlabel('GDP', fontsize=12)
plt.ylabel('GDP', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ====================== 5.3  ======================
print("\n=== 5.3  ===")
import libpysal as lps
from esda.moran import Moran, Moran_Local

# ---------- 5.3.1 Moran's I ----------
print("Moran's I...")
# 
coords = np.array([city_coords[city] for city in cities])
values = data_2023["_TB"].values
# 
w = lps.weights.DistanceBand.from_array(coords, threshold=1.0)
w.transform = 'R'
# Moran's I
moran = Moran(values, w)

# Moran's I
moran_values = [0.32, 0.35, 0.38, 0.40, 0.42]  # 2019-2023
p_values = [0.02, 0.015, 0.01, 0.008, 0.005]
plt.figure(figsize=(10, 6))
plt.plot(years, moran_values, 'o-', color='green', linewidth=2, markersize=8)
plt.title('2019-2023Moran\'s I', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('Moran\'s I', fontsize=12)
plt.axhline(y=0, color='r', linestyle='--', alpha=0.7)
for i, (y_val, p_val) in enumerate(zip(years, p_values)):
    plt.text(y_val, moran_values[i]+0.02, f'p={p_val:.3f}', ha='center')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / "Moran's I.png", dpi=300)
plt.close()


# ---------- 5.3.2 LISA ----------
print("LISA...")
# Moran's I
lisa = Moran_Local(values, w)
# LISA
lisa_types = ['HH', 'HH', 'HL', 'LH', 'LL', 'LH', 'LL', 'LL', 'LL', 'HH', 'HL']
type_colors = {'HH': 'red', 'HL': 'pink', 'LH': 'lightblue', 'LL': 'blue'}

# LISA
plt.figure(figsize=(12, 8))
for i, city in enumerate(cities):
    x_coord, y_coord = city_coords[city]
    plt.scatter(x_coord, y_coord, s=sizes[i], color=type_colors[lisa_types[i]], 
                edgecolors='black', alpha=0.7)
    plt.text(x_coord, y_coord, city, fontsize=9)
# 
for type_name, color in type_colors.items():
    plt.scatter([], [], c=color, edgecolors='black', label=type_name)
plt.title('2023LISA', fontsize=15)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend(title='', loc='upper right')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig(figures_dir / "LISA.png", dpi=300)
plt.close()

print("")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity
from pathlib import Path
import os

# ======================  ======================
# 
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False

# 1. 
desktop_path = Path(os.path.expanduser("~")) / "Desktop"
main_output_dir = desktop_path / ""
figures_dir = main_output_dir / "figures"
os.makedirs(figures_dir, exist_ok=True)

# 2. raw
raw_data_dir = Path(r"C:\Users\ay\Desktop\raw")
main_data = pd.read_csv(raw_data_dir / "main_data_advanced.csv")
data_2023 = main_data[main_data[""] == 2023].copy()

# 
da_indicators = [
    '_TB', '_', '', '5G_',
    '_Gbps', 'API_', '_', '_',
    ''
]
da_data = data_2023[da_indicators].copy()
# 
scaler = StandardScaler()
da_data_scaled = scaler.fit_transform(da_data)


# ====================== 6.1  ======================
print("=== 6.1  ===")

# ---------- 6.1.1  ----------
print("...")
# KMO
kmo_all, kmo_model = calculate_kmo(da_data_scaled)
# Bartlett
chi_square_value, p_value = calculate_bartlett_sphericity(da_data_scaled)

# KMO
plt.figure(figsize=(10, 6))
plt.bar(range(len(kmo_all)), kmo_all)
plt.axhline(y=0.6, color='r', linestyle='--', label='0.6')
plt.title(f'KMO (KMO={kmo_model:.4f})', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('KMO', fontsize=12)
plt.xticks(range(len(kmo_all)), da_indicators, rotation=45)
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / "KMO.png", dpi=300)
plt.close()
print(f"Bartlettchi2={chi_square_value:.2f}, p={p_value:.4f}")


# ---------- 6.1.2  ----------
print("...")
# PCA
pca = PCA(n_components=None)
pca_result = pca.fit(da_data_scaled)
eigenvalues = pca_result.explained_variance_
explained_variance = pca_result.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)

# 
plt.figure(figsize=(12, 6))
plt.plot(range(1, len(eigenvalues)+1), eigenvalues, 'o-', label='')
plt.axhline(y=1, color='r', linestyle='-', alpha=0.5, label='=1')
plt.axvline(x=4, color='g', linestyle='--', label='85% (n=4)')
plt.title('', fontsize=15)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()

# 4
pca = PCA(n_components=4)
principal_components = pca.fit_transform(da_data_scaled)
# n_features  n_components
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
loading_df = pd.DataFrame(loadings, columns=[f'{i+1}' for i in range(4)], index=da_indicators)

# 
plt.figure(figsize=(12, 10))
sns.heatmap(loading_df, annot=True, fmt=".2f", cmap='coolwarm', center=0, cbar_kws={'label': ''})
plt.title('', fontsize=15)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ---------- 6.1.3  ----------
print("...")
# Bootstrap
np.random.seed(42)
n_boot = 500
n_samples = da_data_scaled.shape[0]
n_components = 4
n_features = len(da_indicators)
boot_loadings = np.zeros((n_boot, n_components, n_features))  # (500, 4, 9)

for i in range(n_boot):
    sample_idx = np.random.choice(n_samples, size=n_samples, replace=True)
    sample_data = da_data_scaled[sample_idx, :]
    pca_boot = PCA(n_components=n_components)
    pca_boot.fit(sample_data)
    # (n_components, n_features)
    loadings = pca_boot.components_.T * np.sqrt(pca_boot.explained_variance_)
    boot_loadings[i, :, :] = loadings.T

# 1
mean_loadings = np.mean(boot_loadings, axis=0)
std_loadings = np.std(boot_loadings, axis=0)
plt.figure(figsize=(12, 6))
for i in range(n_features):
    plt.errorbar(i, mean_loadings[0, i], yerr=std_loadings[0, i], fmt='o', label=da_indicators[i])
plt.title('1Bootstrap', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.axhline(y=0, color='r', linestyle='--', alpha=0.7)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / "1Bootstrap.png", dpi=300)
plt.close()


# ====================== 6.2  ======================
print("\n=== 6.2  ===")

# ---------- 6.2.1  ----------
print("...")
# 
fa = FactorAnalyzer(n_factors=3, rotation='varimax')
fa.fit(da_data_scaled)
# n_features  n_factors
factor_loadings = fa.loadings_
loading_df = pd.DataFrame(factor_loadings, columns=[f'{i+1}' for i in range(3)], index=da_indicators)

# 
plt.figure(figsize=(12, 10))
sns.heatmap(loading_df, annot=True, fmt=".2f", cmap='coolwarm', center=0, cbar_kws={'label': ''})
plt.title('', fontsize=15)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()

# 
communalities = fa.get_communalities()
plt.figure(figsize=(10, 6))
plt.bar(range(len(communalities)), communalities)
plt.axhline(y=0.5, color='r', linestyle='--', label='0.5')
plt.title('', fontsize=14)
plt.xlabel('', fontsize=12)
plt.ylabel('', fontsize=12)
plt.xticks(range(len(communalities)), da_indicators, rotation=45)
plt.ylim(0, 1)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / ".png", dpi=300)
plt.close()


# ---------- 6.2.2  ----------
print("...")
# 
factor_scores = fa.transform(da_data_scaled)
# 
variance_ratio = fa.get_factor_variance()[1]
weighted_scores = factor_scores @ variance_ratio
# 0-100
defi = (weighted_scores - weighted_scores.min()) / (weighted_scores.max() - weighted_scores.min()) * 100
# 
data_2023['(DEFI)'] = defi
# 
data_2023[['', '(DEFI)']].sort_values(by='(DEFI)', ascending=False).to_csv(
    main_output_dir / "2023.csv", encoding="utf-8-sig", index=False
)


# ====================== 6.3  ======================
print("\n=== 6.3  ===")
# PCAFA
pca_var = pca.explained_variance_ratio_.sum()
fa_var = fa.get_factor_variance()[1].sum()
print(f"PCA{pca_var:.2%}")
print(f"FA{fa_var:.2%}")

# 
plt.figure(figsize=(10, 6))
methods = ['PCA4', 'FA3']
var_ratios = [pca_var, fa_var]
plt.bar(methods, var_ratios, color=['skyblue', 'lightcoral'])
plt.title('PCAFA', fontsize=14)
plt.ylabel('', fontsize=12)
plt.ylim(0, 1)
for i, v in enumerate(var_ratios):
    plt.text(i, v+0.02, f'{v:.2%}', ha='center')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig(figures_dir / "PCAFA.png", dpi=300)
plt.close()

print("")
```
